```python
import numpy as np
import pandas as pd
Lung_cancer_MG = pd.read_csv('LungCancer/edges_df_with_relative_abundance_MG.csv', sep = ',')
Lung_cancer_MT = pd.read_csv('LungCancer/edges_df_with_RL_LC_MT_HostT.csv', sep = '\t')

COPD_BAL_MG = pd.read_csv('COPD_Imran/edges_df_with_RL_COPD_patients_MG_HostT.csv', sep = '\t')
COPD_BAL_MT = pd.read_csv('COPD_Imran/edges_df_with_RL_COPD_patients_MT_HostT.csv', sep = '\t')

COPD_BAL_MG_ctl = pd.read_csv('COPD_Imran/edges_df_with_RL_SC_MG_HostT.csv', sep = '\t')
COPD_BAL_MT_ctl = pd.read_csv('COPD_Imran/edges_df_with_RL_SC_MT_HostT.csv', sep = '\t')

COPD_spu_MG = pd.read_csv('COPD_Yan/edges_df_with_RL_COPD_MG_HostT.csv', sep = '\t')
COPD_spu_MG_ctl = pd.read_csv('COPD_Yan/edges_df_with_RL_15SC_MG_HostT.csv', sep = '\t')
```

```python

# Store the DataFrames in a list with labels
clusters = {
    'Lung_cancer_MG': Lung_cancer_MG,
    'Lung_cancer_MT': Lung_cancer_MT,
    'COPD_BAL_MG': COPD_BAL_MG,
    'COPD_BAL_MT': COPD_BAL_MT,
    'COPD_BAL_MG_ctl': COPD_BAL_MG_ctl,
    'COPD_BAL_MT_ctl': COPD_BAL_MT_ctl,
    'COPD_spu_MG': COPD_spu_MG,
    'COPD_spu_MG_ctl': COPD_spu_MG_ctl
}

# Prepare a list to store the results
results = []

for name, df in clusters.items():
    data = df['weight']
    median = np.median(data)
    q1 = np.percentile(data, 25)
    q3 = np.percentile(data, 75)
    iqr = q3 - q1

    results.append({
        'group': name,
        'Median': median,
#         'Q1': q1,
#         'Q3': q3,
        'IQR': iqr
    })

# Convert to a new DataFrame
summary_df = pd.DataFrame(results)

summary_df
```
